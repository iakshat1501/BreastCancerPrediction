```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


csv_file = "Cancer_data.csv"
df = pd.read_csv('Cancer_Data.csv')
df.drop("Unnamed: 32", axis=1, inplace=True)


df["Diagnosis Found"] = df["Diagnosis Found"].map({"M": 1, "B": 0})


columns_to_exclude = ["Cancer_id", "Diagnosis Found"]
correlation_matrix = df.drop(columns=columns_to_exclude).corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".1f")
plt.title("Correlation Heatmap (All Columns except 'Cancer_id' and 'Diagnosis Found')")
plt.show()


diagnosis_M = df[df["Diagnosis Found"] == 1]
diagnosis_B = df[df["Diagnosis Found"] == 0]

for column_name in df.columns:
    if column_name not in columns_to_exclude:
        plt.figure(figsize=(8, 6))
        sns.histplot(diagnosis_M[column_name], color="red", label="M", alpha=0.5)
        sns.histplot(diagnosis_B[column_name], color="blue", label="B", alpha=0.5)
        plt.xlabel(column_name)
        plt.ylabel("Frequency")
        plt.title(f"{column_name} distribution for 'M' vs. 'B' diagnoses")
        plt.legend()
        plt.show()
```

**Transforming out data for model**
**X represents the independent variable(feature variable) and Y represents the dependent variable(target variable) in our dataset.**
**Finding its accuracy**

```python
X = df.drop(columns=['Cancer_id', 'Diagnosis Found'])
Y = df['Diagnosis Found']
print(X.shape)
print(Y.shape)
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

#Standardisation
ss = StandardScaler()
X_train=ss.fit_transform(X_train)
X_test= ss.fit_transform(X_test)
X_train
X_test
Y_train
Y_test
```

```python
# Before splitting
print("Shape of X-->",X.shape )
print("Shape of Y-->",Y.shape )
```

```python
# After splitting
print("Shape of X_train-->",X_train.shape )
print("Shape of X_test-->",X_test.shape )
print("Shape of Y_train-->",Y_train.shape )
print("Shape of Y_test-->",Y_test.shape )
```

```python
from sklearn.linear_model import LogisticRegression
lrc=LogisticRegression()
lrc.fit(X_train,Y_train)
Y1_pred=lrc.predict(X_test)
print(" Prediction: \n",Y1_pred)
from sklearn.metrics import confusion_matrix
cm1=confusion_matrix(Y_test,Y1_pred)
print("\n Confusion Matrix: \n",cm1)
#Complete performance report
from sklearn.metrics import classification_report
print("\n Classification report of Logistic Regression Model: \n",classification_report(Y_test,Y1_pred))
```

```python
import matplotlib.pyplot as plt

# Plot the predicted values
plt.plot(Y1_pred, marker='o', linestyle='-')
plt.xlabel('Index')
plt.ylabel('Predicted Values')
plt.title('Predicted Values Plot')
plt.grid(True)
plt.show()
```

```python
#Model Creation ---SUPPORT VECTOR MACHINE --- classification (SVC)
#DEFAULT KRNEL=RBF OR GAUSSIAN
from sklearn.svm import SVC
svmc=SVC(kernel='rbf')
svmc.fit(X_train, Y_train)
#Prediction -- on test dataset -- X_test
Y2_pred=svmc.predict(X_test)
print(" Prediction: \n",Y2_pred)
#Evaluate performance of model --- Confusion Matrix
from sklearn.metrics import confusion_matrix
cm2=confusion_matrix(Y_test,Y2_pred)
print("\n Confusion Matrix: \n",cm2)
#Check the accuracy
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y2_pred)*100
#Complete performance report
from sklearn.metrics import classification_report
print("\n Classification report of Support Vector Machine(rbf): \n",classification_report(Y_test,Y2_pred))
```
